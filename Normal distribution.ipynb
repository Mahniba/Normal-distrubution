```python
#imports
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from scipy import stats
```

```python
#parameters
mean = 0 #represents center of distribution

std_dev = 1 #indicates spread of data
#generate data points

x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000) #range of values, array with range of values ranging from (u-4q) to (u-4q)
#this range captures most of the distribution(99.7% of data in a normal distribution)

y = (1 / (std_dev *np.sqrt(2 *np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) **2) #normal distribution formula
#calculates the probability density of the normal distribution, the formula is : f(x) =(1/(qsqrt(2pi)))*e^(-(1/2)((x-u)q)^2)

#plotting the normal distribution
plt.figure(figsize =(10,6)) #set figure size for better visibility

#plot PDF curve in blue(probability density function)
plt.plot(x, y, label = 'Normal Distribution', color = 'blue')

#set title with mean and standard deviation
plt.title('Normal Distribution (Mean ={}, Std Dev ={})'.format(mean,std_dev))

#label for y axis
plt.xlabel('X-axis')

#label for x axis
plt.ylabel('Probability Density')

#verical line at mean, shown in red
plt.axvline(mean, color ='red', linestyle='--', label = 'Mean')

#vertical line at mean + q shown in green left boundary
plt.axvline(mean + std_dev, color='green',linestyle='--', label='Mean + Std Dev')

#vertical line at mean - q, also shown in green right boundary
plt.axvline(mean - std_dev, color = 'green', linestyle = '--', label='Mean - Std Dev')

#show legend to identify lines on the plot
plt.legend()

#add a grid for better readability
plt.grid()

#display plot
plt.show()
```

```python
def calculate_normal_distribution(mean, std_dev, x):
    """Calculate the probability density function of a normal distribution."""
    return (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# User input for parameters
mean = float(input("Enter the mean (): "))
std_dev = float(input("Enter the standard deviation (): "))
x = float(input("Enter the value to calculate P(X=x): "))

# Calculate probability density at x
probability_density = calculate_normal_distribution(mean, std_dev, x)

print(f"The probability density at x={x} for N(={mean}, ={std_dev}) is {probability_density:.4f}")
```

```python
# Parameters
mean = 65
std_dev = 9

# Probability calculations
p_less_than_54 = norm.cdf(54, mean, std_dev)
p_at_least_80 = 1 - norm.cdf(80, mean, std_dev)
p_between_70_86 = norm.cdf(86, mean, std_dev) - norm.cdf(70, mean, std_dev)

print(f"P(X < 54): {p_less_than_54:.4f}")
print(f"P(X  80): {p_at_least_80:.4f}")
print(f"P(70 < X < 86): {p_between_70_86:.4f}")

# Plotting the normal distribution
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
y = norm.pdf(x, mean, std_dev)

plt.plot(x, y)
plt.title('Normal Distribution (mean=65, std dev=9)')
plt.xlabel('Scores')
plt.ylabel('Probability Density')
plt.grid()
plt.show()
```

```python
def normal_distribution(mu, sigma, lower_bound, upper_bound):
    """
    Calculate the probability of a value falling between two bounds in a normal distribution.

    Parameters:
    mu (float): The mean of the normal distribution.
    sigma (float): The standard deviation of the normal distribution.
    lower_bound (float): The lower bound of the range.
    upper_bound (float): The upper bound of the range.

    Returns:
    float: The probability that a value falls between lower_bound and upper_bound.
    """
    
    # Calculate the cumulative distribution function (CDF) at the lower bound
    cdf_lower = stats.norm.cdf(lower_bound, mu, sigma)
    
    # Calculate the cumulative distribution function (CDF) at the upper bound
    cdf_upper = stats.norm.cdf(upper_bound, mu, sigma)
    
    # The probability of falling between the bounds is the difference of the CDF values
    return cdf_upper - cdf_lower

# Example usage
mean = 0 # Mean of the normal distribution (for standard normal distribution)
std_dev = 1 # Standard deviation of the normal distribution (for standard normal distribution)
range_start = -1 # Lower bound for the probability calculation
range_end = 1 # Upper bound for the probability calculation

# Call the function to calculate the probability between range_start and range_end
probability = normal_distribution(mean, std_dev, range_start, range_end)

# Print the calculated probability
print(f"Probability between {range_start} and {range_end}: {probability}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data, mean, std_dev):
    """
    Calculate the z-scores for a given dataset.

    Parameters:
    data (array-like): The dataset for which to calculate z-scores.
    mean (float): The mean of the dataset.
    std_dev (float): The standard deviation of the dataset.

    Returns:
    np.array: An array of z-scores corresponding to the input data.
    """
    return (data - mean) / std_dev

# Example parameters
mean = float(input("enter the mean of the distribution:")) # Mean of the distribution
std_dev = float(input("enter the standard deviation of the distribution:")) # Standard deviation of the distribution

# Generate a range of x values around the mean
x_values = float(input("enter the x value to calculate the z_score of the distribution:"))

# Calculate z-scores for these x values
z_scores = calculate_z_scores(x_values, mean, std_dev)

# Displaying the z-scores
print(f"the z-score for x ={x_values} is: {z_scores: .2f}")

# Plotting the z-scores
plt.figure(figsize=(10, 5))
plt.plot(x_values, z_scores, label='Z-Scores', color='blue')
plt.axhline(0, color='red', linestyle='--', label='Z = 0 (Mean)')
plt.title('Z-Scores for Given Mean and Standard Deviation')
plt.xlabel('X Values')
plt.ylabel('Z-Scores')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

def calculate_z_score(x, mean, std_dev):
    """
    Calculate the z-score for a given x value.

    Parameters:
    x (float): The x value for which to calculate the z-score.
    mean (float): The mean of the distribution.
    std_dev (float): The standard deviation of the distribution.

    Returns:
    float: The z-score corresponding to the input x value.
    """
    return (x - mean) / std_dev

def get_cdf_value(z):
    """
    Get the cumulative distribution function (CDF) value for a given z-score.

    Parameters:
    z (float): The z-score.

    Returns:
    float: The CDF value corresponding to the z-score.
    """
    return stats.norm.cdf(z)

# User input for mean, standard deviation, and x value
mean = float(input("Enter the mean of the distribution: "))
std_dev = float(input("Enter the standard deviation of the distribution: "))
x_value = float(input("Enter the x value to calculate the z-score: "))

# Calculate z-score
z_score = calculate_z_score(x_value, mean, std_dev)

# Get CDF value for the calculated z-score
cdf_value = get_cdf_value(z_score)

# Displaying results
print(f"The z-score for x = {x_value} is: {z_score:.4f}")
print(f"The cumulative probability (CDF) corresponding to z = {z_score:.4f} is: {cdf_value:.4f}")
```
