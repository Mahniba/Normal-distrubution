```python
import numpy as np
import matplotlib.pyplot as plt
from math import comb

#FUNCTION FOR BINOMIAL DESTRIBUTION
def binomial_destribution():

    #promp user for input ( number of trials)
    n = int(input("Enter the Number of Trials (n): "))

    #asking the user to enter the probability of success
    p = float(input("Enter the probability of success(p): "))

    if p<0 or p>1:
        print("Probability of sucess p must be between 0 and 1, Please try again")
        return
    
    #Calculate the probabilities for all possible success (0 to n)
    x = np.arange(0, n+1)
    probabilities = [comb(n, k) * (p**k)*((1-p)**(n-k)) for k in x]

    #Plot the histogram
    plt.bar(x, probabilities, color='skyblue', edgecolor = 'black')
    plt.title(f'Binomial Destribution (n = {n}, p = {p})')
    plt.xlabel("Number of Successes")
    plt.ylabel("Probability")
    plt.xticks(x)
    plt.grid(axis='y', linestyle = '--', alpha=0.7)
    plt.show()

#Run the Binomial destribution function
if __name__ == "__main__":
    binomial_destribution()


    
```
